```python
def createGraph(V, edges):
    graph = [[0 for _ in range(V)]for _ in range(V)]

    for edge in edges:
        u = edge[0]
        v = edge[1]

        # for an edge, mark 1 in the adjacency matrix
        graph[u][v] = 1
        #graph[v][u] = v

        #for undirected graph:
        graph[v][u] = 1
    return graph
```

```python
if __name__ == "__main__":
    V = 4
    #edges = [[0,1],[0,2],[2,3],[1,2]]
    edges = [[0,1], [0,2], [2,3]]
    graph = createGraph(V, edges)

    for i in range(V):
        for j in range(V):
            print(graph[i][j], end=" ")
        print()
```

```python
def BFS(V, s, graph):
    color = ["White"] * V
    distance = [-1] * V
    parent = [-1] * V
    for i in range(V):
        if i != s:
            color[i] = "White"
            distance[i] = -1
            parent[i] = -1
    color[s] = "Grey"
    parent[s] = -1
    distance[s] = 0
    print(s)
    
    q = []
    q.append(s)

    while q:
        u = q.pop(0)
        for i in range(len(graph[u])):
            if graph[u][i] == 1:
                if color[i] == "White":
                    color[i] = "Grey"
                    distance[i] = distance[u] + 1
                    parent[i] = u
                    print(i)
                    q.append(i)
        color[u] = "Black"
```
