```python
def createGraph(noOfVertices, edges):
    graph = [[0 for _ in range(noOfVertices)] for _ in range(noOfVertices)]

    for edge in edges:
        u = edge[0]-1
        v = edge[1]-1

        graph[u][v] = 1

        #if the graph is undirected, use graph[v][u] = 1

    return graph
```

```python
V = 6
edges = [[1,2],[1,4],[2,5],[3,5],[3,6],[4,2],[5,4],[6,6]]
graph = createGraph(V, edges)
```

```python
def dfsForCycle(graph, V):
    visited = [False]*V
    stackOfVisited = [False]*V

    for i in range(V):
        if not visited[i]:
            if hasCycle(graph, V, visited, stackOfVisited, i):
                return True

    return False
```

```python
def hasCycle(graph, V, visited, stackOfVisited, s):
    #if the node is in the stack of maintained nodes and if it's already marked true/visited, it means, it forms a cycle now.
    if stackOfVisited[s]:
        return True

    #if the node is already visited, obviously it's tree is visited and cycle was not found
    if visited[s]:
        return False

    visited[s] = True
    stackOfVisited[s] = True
    
    for i in range(V):
        if graph[s][i] == 1:
            if hasCycle(graph, V, visited, stackOfVisited, i):
                return True

    stackOfVisited[s] = False
    return False
```

```python
dfsForCycle(graph, V)
```
