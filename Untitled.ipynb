```python
from collections import deque
```

```python
def createGraph(V, edges):
    graph = [[0 for _ in range(V)] for _ in range(V)]

    for edge in edges:
        u = edge[0]
        v = edge[1]
        graph[u][v] = 1

        #if the graph is undirected:
        graph[v][u] = 1

    return graph
```

```python
if __name__ == "__main__":
    V = 4
    edges = [[0,1],[0,2],[2,3]]

    graph = createGraph(V, edges)

    print("Adjacency Matrix Representation:")
    for i in range(V):
        for j in range(V):
            print(graph[i][j], end=" ")
        print()
```

```python
def bfs(start, graph, visited, parent):
    #q = []
    q = deque()

    #q.append(start)
    q.append([start, -1])
    visited[start] = True
    #parent[start] = -1

    while q:
        #node = q.pop(0)
        node = q[0][0]
        parent = q[0][1]
        q.popleft()
        
        for neighbor in graph[node]:
            if not visited[neighbor]:
                q.append([neighbor,node])
                visited[neighbor] = True
                #parent[neighbor] = node

            elif neighbor != parent:
                print("cycle detected here",neighbor,parent)
                return True
        #print(visited)
        #print(parent)
    return False
```

```python
def isCycle(graph):
    V = len(graph)
    visited = [False] * V
    parent = [-1] * V

    for i in range(V):
        if not visited[i]:
            if bfs(i, graph, visited, parent):
                return True
    return False
```
